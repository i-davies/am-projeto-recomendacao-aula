## Visualização Simples

Vamos visualizar a distribuição de algumas colunas interessantes:

- **Danceability**: Mede o quão adequada uma música é para dançar (0.0 a 1.0).
- **Energy**: Mede a intensidade e atividade da música (0.0 a 1.0).

```python
plt.figure(figsize=(10, 6))

# Gráfico de Danceability
plt.hist(df['danceability'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel('Danceability')
plt.ylabel('Frequência')
plt.title('Distribuição da Dançabilidade')
plt.show()
```

### Relação entre Energia e Barulho (Loudness)

Será que músicas mais enérgicas são mais barulhentas?

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['energy'], df['loudness'], alpha=0.5)
plt.xlabel('Energy')
plt.ylabel('Loudness')
plt.title('Energia vs. Loudness')
plt.show()
```

### Músicas "Energéticas" mas "Silenciosas"?

No gráfico acima, podemos ver que geralmente Energia e Loudness crescem juntos. Mas existem exceções?
Vamos buscar músicas com **Alta Energia** (> 0.8) mas **Loudness Baixo** (< -10 dB). Essas músicas podem confundir nosso Perceptron simples se não tratarmos os dados corretamente.

```python
# Filtrando exceções: Alta Energia e Baixo Volume
excecoes = df[(df['energy'] > 0.8) & (df['loudness'] < -10)]

# Ordenando pelas mais energéticas
excecoes_ordenadas = excecoes[['track_id','track_name', 'artists', 'energy', 'loudness']].sort_values(by='energy', ascending=False)

# Exibindo as top 10
excecoes_ordenadas.head(10)
```

## Pesquisando no Dataset

Muitas vezes precisamos encontrar itens específicos no nosso conjunto de dados. O Pandas nos permite fazer isso de forma muito eficiente usando filtros.

Abaixo, vamos ver como pesquisar por:
- **Artista**
- **Nome da Música**
- **ID da Faixa (track_id)**

```python
# 1. Pesquisando por um artista específico (ex: 'Ed Sheeran')
# Usamos str.contains para buscar partes do nome. 
# case=False ignora se está maiúsculo ou minúsculo.
# na=False garante que valores nulos não quebrem a busca.
busca_artista = df[df['artists'].str.contains('Ed Sheeran', case=False, na=False)]
busca_artista.head()
```

```python
# 2. Pesquisando por nome da música (ex: musicas com 'Love' no nome)
busca_musica = df[df['track_name'].str.contains('Love', case=False, na=False)]
busca_musica.head()
```

```python
# 3. Pesquisando por um track_id específico
# Aqui usamos a igualdade direta (==) pois queremos o ID exato
id_para_buscar = '5SuOikwiRyPMVoIQDJUgSV' # Exemplo de ID (Gen Hoshino - Comedy)
busca_id = df[df['track_id'] == id_para_buscar]
busca_id
```
